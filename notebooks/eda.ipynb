## Exploratory Data Analysis  Key Insights

### Dataset Overview
- The dataset consists of **1,200 employee records** with **28 features**.
- Out of these, **9 features are categorical** and the remaining are numerical.
- The dataset is sufficiently large and balanced for statistical analysis and machine learning modeling.


### Gender Distribution
- **60.4% of employees are Male** and **39.6% are Female**.
- The workforce shows moderate gender imbalance, which may influence department-level performance trends and satisfaction metrics.

### Educational Background
- Majority of employees come from **Life Sciences (41%)** and **Medical (32%)** backgrounds.
- This indicates a strong domain-focused workforce, especially suited for R&D and development-oriented roles.


### Marital Status
- **45.7% of employees are Married**
- **32% are Single**
- **22.3% are Divorced**

Marital status may influence factors such as work-life balance, overtime preferences, and travel frequency.

### Department-wise Distribution
The top three departments by employee count are:
1. **Sales  31.1%**
2. **Development  30%**
3. **R&D  28.6%**

These departments together form nearly **90% of the workforce**, making them critical focus areas for performance improvement initiatives.


### Business Travel
- **18.5% of employees travel frequently** for business purposes.
- Frequent travel can impact employee fatigue, job satisfaction, and long-term performance.


### Overtime
- **29.4% of employees work overtime**.
- This suggests potential workload imbalance, which could affect employee performance and burnout levels.


### Attrition
- **85.2% of employees show no attrition**, indicating strong employee retention.
- Low attrition allows the company to focus more on performance optimization rather than replacement hiring.

## EDA Summary  Business Perspective

- The workforce is dominated by Sales, Development, and R&D departments, making them primary targets for performance analysis.
- Employee performance is likely influenced by workload (overtime), satisfaction factors, and career progression rather than attrition.
- The dataset shows strong stability, making it suitable for building a reliable performance prediction model.
