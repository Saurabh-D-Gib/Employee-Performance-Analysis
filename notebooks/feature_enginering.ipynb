```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("D:\ML Projects\Employee Performance Analysis\data\INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xlsx")
```

```python
df.drop(columns=['EmpNumber'], inplace=True)
```

```python
df['OverTime']=df['OverTime'].map({'Yes':1, 'No':0})
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
df['Gender']=df['Gender'].map({'Male':1, 'Female':0})
```

```python
df.head()
```

Added Feature via engieering

```python
df['ExperienceStability'] = (df['ExperienceYearsAtThisCompany'] /(df['TotalWorkExperienceInYears'] + 1))
```

```python
df['PromotionStagnation'] = (df['YearsSinceLastPromotion'] /(df['TotalWorkExperienceInYears'] + 1))
```

```python
df['WorkStressIndex'] = df['OverTime'] * (5 - df['EmpWorkLifeBalance'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    if col not in ['OverTime', 'Attrition']:
        df[col] = le.fit_transform(df[col])
```

```python
df.to_pickle(r"D:\ML Projects\Employee Performance Analysis\artifacts\encoders\labelencoder.pkl")
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    drop='first',
    sparse_output=False,
    handle_unknown='ignore'
)

encoded_array = ohe.fit_transform(df[cat_cols])

encoded_df = pd.DataFrame(
    encoded_array,
    columns=ohe.get_feature_names_out(cat_cols),
    index=df.index
)

df = df.drop(columns=cat_cols)
df = pd.concat([df, encoded_df], axis=1)
```
